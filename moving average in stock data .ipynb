```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
stock=yf.download(['SBIN.NS'],period='3mo')
```

```python
stock.head(5)
```

```python
stock['10MA']=stock[['Close']].rolling(10).mean()
stock['30MA']=stock[['Close']].rolling(30).mean()
```

```python
stock
```

```python
stock.dropna(inplace=True)
```

```python
plt.plot(stock['Close'],label='Close')
plt.plot(stock['10MA'],label='10MA')
plt.plot(stock['30MA'],label='30MA')

plt.legend(loc='upper right')
```
